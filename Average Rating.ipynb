```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName('Average Quiz')
sc = SparkContext.getOrCreate(conf)
```

```
rdd = sc.textFile('/FileStore/tables/Average_Quiz.txt')
```

```
rdd.collect()
```

```
### split the list into tuples and change data type
rdd1 = rdd.map(lambda x: (x.split(',')[0], x.split(',')[1], float(x.split(',')[2]), 1))
rdd1.collect()
```

```
## convert the tuples into key valye pairs
rdd2 = rdd1.map(lambda x: (x[0], (x[2], x[3])))
```

```
rdd2.collect()
```

```
rdd3 = rdd2.reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
rdd3.collect()
```

```
rdd4 = rdd3.map(lambda x: (x[0], x[1][0]/ x[1][1]))
rdd4.collect()
```

```
## MIN AND MAX VALUE USING REDUCE BY KEY

rdd.reduceBykey(lambda x, y: x if x < y else y) ## min value
rdd.reduceBykey(lambda x, y: x if x > y else y) ## max value
```
